```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Reading The Image

```python
img=cv2.imread('opencv_image.png',1)
```

### Converting the Color Format From BGR to RGB format

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### 2D Convolution Function

```python
# This will reduce the noises at the Edge and will give you a clear Image
# Syntax For Kernel Of 2D Convolution Method
# kernel array divided by multiplication of the kernel size
# in my case kernel size=(4,4) and the shape of the kernel is (4,4)
# So the (4,4) Kernel is divided by multiplication of its shape (4*4)
kernel=np.ones((4,4),np.float32)/(4*4) #it is divided by the multiplication of kernel shape
```

```python
dest=cv2.filter2D(img,10,kernel)
```

### Blur Method

```python
blur_img=cv2.blur(img,(5,5))
```

### Gaussian Blur Method

```python
## In the Guassian Method, the value present in center of the Kernel has the higher weightage
```

```python
gau_blur_image=cv2.GaussianBlur(img,(5,5),0)
```

```python
titles=['image','2d-Convolution','Blur_Image','Gaussian_Image']
images=[img,dest,blur_img,gau_blur_image]
```

### Displaying The Image Using Matplotlib

```python
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
