```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Reading Image

```python
img=cv2.imread('Water Drop.jfif',1)
```

### Change The Color Format From BGR to RGB Format

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Median Filter

```python
median=cv2.medianBlur(img,5) # Here the kernel size should be in odd numbers
# Higher The Kernel Size To Get A Best Result
```

### Bilateral Filter

```python
bilat_img=cv2.bilateralFilter(img,50,75,75)
#cv2.bilateralFilter(image_name,diameter of kernel,sigma color,sigma Space)
# if you want to blur the image and preserve the Edges to be Sharp then we can use this
```

```python
images=[img,median,bilat_img]
titles=['Original Image','Median Image','Bilateral Image']
```

```python
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
