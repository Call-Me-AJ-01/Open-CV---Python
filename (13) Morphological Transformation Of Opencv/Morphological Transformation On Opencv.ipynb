### Dilation Method

```python
img=cv2.imread('balls.jfif',0)
success,mask=cv2.threshold(img,220,225,cv2.THRESH_BINARY)
```

```python
## KERNEL is normally a shape that we want to apply on the image
```

```python
kernel=np.ones((2,2),np.uint8)
#print(kernel)
## Herein if u increase the size of Ones, Then those dots will be removed but size of the image will be little bigger
```

```python
dilation=cv2.dilate(mask,kernel,iterations=1) 
# If Any Pixel Of The Kernel Is '1' The All The Pixel Is Considered as '1'
# Iteration is number of times the Dilation needs to be performed on the Image
# So when u run dilation also there may be some dots inside the picture, to remove it completely we 
# set to perform it in Loop...DEFAULT value for Iteration is "1
images=[img,mask,dilation]
titles=['Image','mask','dilation']

for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
plt.show()
```

### Erosion Method

```python
erosion=cv2.erode(mask,kernel) 
# If All The Pixel Of The Kernel Is '1' Them All The Pixel Value Is Considered As '1'
# The Size will not here
```

### Morphology Functions

```python
open_=cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)# MORPH_OPEN 1st erosion is done and 2nd dilation is done
close_=cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)# MORPH_CLOSE 1st dilation is done and 2nd erosion is done
```

```python
images=[img,mask,dilation,erosion,open_,close_]
titles=['Image','Mask','Dilation','Erosion','Morph_open','Morph_close'] 
```

### Using Matplotlib to Show All The Images

```python
for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

```python
cv2.imshow('img',img)
cv2.imshow('mask',mask)
cv2.imshow('dilation',dilation)
cv2.imshow('erosion',erosion)
cv2.imshow('open_',open_)
cv2.imshow('close_',close_)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
