```python
import cv2
import numpy as np
```

```python
# Creating a Function For Trackbar
def aj(x):
    #print(x)
    pass

#Creating a named Window For Trackbar    
cv2.namedWindow('Tracking')

#Creating The Trackbar

#Minimum Range Trackbar
cv2.createTrackbar('LB','Tracking',0,255,aj)
cv2.createTrackbar('LG','Tracking',0,255,aj)
cv2.createTrackbar('LR','Tracking',0,255,aj)
#Maximum Range Trackbar
cv2.createTrackbar('HB','Tracking',255,255,aj)
cv2.createTrackbar('HG','Tracking',255,255,aj)
cv2.createTrackbar('HR','Tracking',255,255,aj)

#Accessing The Camera
video=cv2.VideoCapture(0)

while True:
    # Reading The Image
    success,img=video.read()
    
    #Cinverting The Image To HSV Format
    hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    
    #Displaying The HSV Converted Image
    cv2.imshow('HSV Image',hsv)
    
    #Displaying The Original Image
    cv2.imshow('Original Image',img)
    
    #Getting The Minimum Range Value From The Trackbar
    l_b=cv2.getTrackbarPos('LB','Tracking')
    l_g=cv2.getTrackbarPos('LG','Tracking')
    l_r=cv2.getTrackbarPos('LR','Tracking')
    
    #Getting The Maximum Range Value From The Trackbar
    h_b=cv2.getTrackbarPos('HB','Tracking')
    h_g=cv2.getTrackbarPos('HG','Tracking')
    h_r=cv2.getTrackbarPos('HR','Tracking')
    
    #In Order To Do Threshold For The Given range In Trackbar in HSV Image
    Min_range=np.array([l_b,l_g,l_r]) # Lower Range Value From The Trackbar
    Max_range=np.array([h_b,h_g,h_r])# Higher Range value From The Trackbar
    
    #Printing The Min and Max Range From The Trackbar
    #print('The Maximum And Minimum Range From The Trackbar')
    #print(Min_range)
    #print(Max_range)
    
    #Thresholding The HSV Image
    mask=cv2.inRange(hsv,Min_range,Max_range)
    
    #Displaying The Threshold Image
    cv2.imshow('Threshold Image',mask)
    
    #Printing The Threshold Array
    #print('\nPrinting The mask Value')
    #print(mask)
    
    #Extracting The Blue Color From The Image
    Extracted_img=cv2.bitwise_and(img,img,mask=mask)
    
    #Printing The Extracted Image Array
    #print('\nPrinting The Extracted Image Value')
    #print(Extracted_img)
    #print()
    
    #Displaying The Extracted Image
    cv2.imshow('Extracted Image',Extracted_img)
    
    if cv2.waitKey(1)==ord('q'):
        break
        
cv2.destroyAllWindows()
```
