```python
import cv2
```

### Reading The Image

```python
colored_img=cv2.imread('opencv_image.png',1)
grayscaled_img=cv2.imread('opencv_image.png',0)
```

### Finding The Threshold Image

```python
ret,th=cv2.threshold(grayscaled_img,200,255,0)
```

### Finding The Contours

```python
contours,hierarchy=cv2.findContours(th,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
```

```python
print('No of Countours :'+str(len(contours)))
```

### Priniting Each Contours

```python
for i in contours:
    print(i)
```

### Drawing The Contours

```python
cv2.drawContours(colored_img,contours,-1,(255,255,255),-1)
```

### Displaying The Image

```python
cv2.imshow('Colored Image',colored_img)
cv2.imshow('Gray Scaled Image',grayscaled_img)
cv2.imshow('th',th)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
