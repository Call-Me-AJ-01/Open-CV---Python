```python
import cv2
import matplotlib.pyplot as plt
```

### Reading Image Using Opencv

```python
img=cv2.imread('Original Image.jfif',1)
```

### Displaying Image Using Opencv

```python
plt.imshow(img)
plt.show()
cv2.imshow('Opencv Window',img)
cv2.waitKey(0)
```

### The Color Change In Matplotlib when Compared To Original Image Is That
        Opencv reads Image In BGR fornat whereas Matplotlib reads Image In RGB format

```python
img_plt=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)# Converting The Color Format From BGR to RGB
```

```python
plt.imshow(img_plt)
plt.show()
cv2.imshow('Opencv Window',img)
cv2.waitKey(0)
```

### Hidding X And Y Values

```python
plt.imshow(img_plt)
plt.xticks([])
plt.yticks([])
plt.show()
```
