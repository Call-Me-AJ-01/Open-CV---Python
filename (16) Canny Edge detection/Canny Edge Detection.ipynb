```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Displaying The Image

```python
img=cv2.imread('Messi.jpg',0)
```

### Converting The Color Format from BGR to RGB Format

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Canny Method

```python
can_img=cv2.Canny(img,100,200)
```

### Displaying The Image Using cv2

```python
cv2.imshow('Canny Image',can_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
images=[img,can_img]
titles=['Original Image','Canny_Image']
```

```python
for i in range(len(images)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
