### Image Gradient
        Image Gradient is a directional Change in the intensity or color in an image

### Image Gradient For Messi Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Reading Image Using Opencv

```python
img=cv2.imread('Messi.jpg',0)
```

### Converting The color Format From BGR to RGB Format

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

### Laplacian Method

```python
lap=cv2.Laplacian(img,cv2.CV_64F)
print('laplacian value before Converting it into absolute value')
print(lap)
cv2.imshow('lap image before converting it into absolute value',lap)
lap=np.uint8(np.absolute(lap))
print('\nlaplacian value after Converting it into absolute value')
print(lap)
cv2.imshow('lap image after converting it into absolute value',lap)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Sobel-X Method or Sobel Gradient

```python
#cv2.sobel(image_name,data_type,sobel_x value,sobel_y value)
#here since we are using sobel_x method we have assigned 1 to sobel_x value
sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0)
print(sobel_x)
print('---------------------------------')
#converting the Negative Value To Positive Value
sobel_x=np.uint8(np.absolute(sobel_x))
print(sobel_x)
```

### Sobel-Y Method or Sobel Gradient

```python
#cv2.sobel(image_name,data_type,sobel_x value,sobel_y value)
#here since we are using sobel_y method we have assigned 1 to sobel_y value
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1)
print(sobel_y)
print('-------------------------------------')
#converting the Negative Value To Positive Value
sobel_x=np.uint8(np.absolute(sobel_y))
print(sobel_y)
```

```python
images=[img,lap,sobel_x,sobel_y]
titles=['Original Images','Laplacian Image','Sobel_x Image','Sobel_y Image']
```

```python
for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
